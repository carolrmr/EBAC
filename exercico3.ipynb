```python
import psycopg2 as pg2 
import pandas as pd
```

```python
conn = pg2.connect(host= 'localhost', port=5432, dbname='dvdrental', user='carolina', password='1234' )
```

```python
cur = conn.cursor()
```

```python
cur.execute('''select  * 
               from film as a
                   left join film_actor  as b
                        on a.film_id  = b.film_id
                   left join actor  as c
                        on b.actor_id  = c.actor_id
                   left join actor as d
                        on c.first_name  = d.first_name 
                   left join actor as e
                        on d.last_name  = e.last_name  
                     limit 100;''')

```

```python
cur.execute('''SELECT 
    a.first_name,
    a.last_name,
    AVG(f.rental_duration) AS avg_rental_duration
FROM 
    actor AS a
JOIN 
    film_actor AS fa ON a.actor_id = fa.actor_id
JOIN 
    film AS f ON fa.film_id = f.film_id
GROUP BY 
    a.first_name, a.last_name;''')
```

```python
conn.commit()
```

```python
data = cur.fetchall()
colnames = [desc.name for desc in cur.description ]
```

```python
cur.close()
conn.close()
```

```python
df= pd.DataFrame(data,columns=colnames)
df.head()
```
