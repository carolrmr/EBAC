## 1) Diferena de peso entre machos e fmeas

Na base de pinguins, podemos ver que h diferena de peso entre machos e fmeas. Mas ser que  uma concluso extrapolvel a toda a populao de pinguins? Assumindo que essa amostra  aleatria e representativa da populao de pinguins, realize um teste T com 5% de significncia para verificar se podemos induzir esta concluso para a populao toda.

```python
from scipy import stats
import seaborn as sns

pg = sns.load_dataset('penguins')

#Pegando os dados de femeas e machos
female_weights = pg[pg['sex'] == 'Female']['body_mass_g']
male_weights = pg[pg['sex'] == 'Male']['body_mass_g']

#Teste t de Student independente
t_statistic, p_value = stats.ttest_ind(female_weights, male_weights)

#Validando estatstica e valor
print(f"T-Estatstica: {t_statistic}")
print(f"P-Valor: {p_value}")

# Verifique se o p-valor  menor que a significncia desejada (0.05)
if p_value < 0.05:
    print("H evidncias para dizer que existe uma diferena significativa nos pesos entre fmeas e machos.")
else:
    print("No h evidncias suficientes para dizer que existe diferena significativa nos pesos entre fmeas e machos.")
```
