## 1) Diferença de peso entre machos e fêmeas

Na base de pinguins, podemos ver que há diferença de peso entre machos e fêmeas. Mas será que é uma conclusão extrapolável a toda a população de pinguins? Assumindo que essa amostra é aleatória e representativa da população de pinguins, realize um teste T com 5% de significância para verificar se podemos induzir esta conclusão para a população toda.

```python
from scipy import stats
import seaborn as sns

pg = sns.load_dataset('penguins')

#Pegando os dados de femeas e machos
female_weights = pg[pg['sex'] == 'Female']['body_mass_g']
male_weights = pg[pg['sex'] == 'Male']['body_mass_g']

#Teste t de Student independente
t_statistic, p_value = stats.ttest_ind(female_weights, male_weights)

#Validando estatística e valor
print(f"T-Estatística: {t_statistic}")
print(f"P-Valor: {p_value}")

# Verifique se o p-valor é menor que a significância desejada (0.05)
if p_value < 0.05:
    print("Há evidências para dizer que existe uma diferença significativa nos pesos entre fêmeas e machos.")
else:
    print("Não há evidências suficientes para dizer que existe diferença significativa nos pesos entre fêmeas e machos.")
```

```python
percentile_5 = pg.groupby('sex')['body_mass_g'].quantile(0.05)

print(percentile_5)
```
