### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
import pandas as pd

sinasc_raw = pd.read_csv('SINASC_RO_2019.csv')

# Calcular a idade média das mães por município
idade_media_maes = sinasc_raw.groupby('munResNome')['IDADEMAE'].mean()

# Calcular a idade média dos pais por município
idade_media_pais = sinasc_raw.groupby('munResNome')['IDADEPAI'].mean()

print("Idade Média das Mães por Município:")
print(idade_media_maes)

print("\nIdade Média dos Pais por Município:")
print(idade_media_pais)
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
import pandas as pd

sinasc_raw = pd.read_csv('SINASC_RO_2019.csv')

sinasc_raw['DTCADASTRO'] = pd.to_datetime(sinasc_raw['DTCADASTRO'])

data_desejada = pd.to_datetime('2019-09-12')
sinasc_filtered = sinasc_raw[sinasc_raw['DTCADASTRO'] == data_desejada]

 #Calcular o peso médio dos bebês por sexo e faixa de escolaridade da mãe
peso_medio_bebes = sinasc_filtered.groupby(['ESCMAE', 'SEXO'])['PESO'].mean()


print("Peso Médio dos Bebês por Sexo e Faixa de Escolaridade da Mãe:")
print(peso_medio_bebes)

```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
import pandas as pd

sinasc_raw = pd.read_csv('SINASC_RO_2019.csv')


bebês_por_município = sinasc_raw.groupby('munResNome').size()

menos_bebês_município = bebês_por_município.idxmin()
sinasc_menor_município = sinasc_raw[sinasc_raw['munResNome'] == menos_bebês_município]

idade_maes_estatisticas = sinasc_menor_município['IDADEMAE'].aggregate(['mean', 'max', 'min'])


idade_pais_estatisticas = sinasc_menor_município['IDADEPAI'].aggregate(['mean', 'max', 'min'])

print("Município com Menos Bebês em 2019:", menos_bebês_município)
print("\nEstatísticas de Idade das Mães:")
print(idade_maes_estatisticas)
print("\nEstatísticas de Idade dos Pais:")
print(idade_pais_estatisticas)
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
import pandas as pd

sinasc_raw = pd.read_csv('SINASC_RO_2019.csv')

sinasc_raw['DTCADASTRO'] = pd.to_datetime(sinasc_raw['DTCADASTRO'])


sinasc_marco = sinasc_raw[sinasc_raw['DTCADASTRO'].dt.month == 3]


bebês_por_município = sinasc_marco.groupby('munResNome').size()

município_mais_bebês = bebês_por_município.idxmax()

sinasc_município_mais_bebês = sinasc_marco[sinasc_marco['munResNome'] == município_mais_bebês]


filhos_vivos_estatísticas = sinasc_município_mais_bebês['QTDFILVIVO'].aggregate(['mean', 'max', 'min'])

idade_pais_estatísticas = sinasc_município_mais_bebês['IDADEPAI'].aggregate(['mean', 'max', 'min'])

print("Município com Mais Bebês no Mês de Março:", município_mais_bebês)
print("Quantidade de Filhos Vivos:")
print(filhos_vivos_estatísticas)
print(" Idade dos Pais:")
print(idade_pais_estatísticas)
```
