### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Carregando os dados usando o seaborn
data = sns.load_dataset('tips')

# Ajustando o modelo de regressão linear usando statsmodels
regression_model = smf.ols('tip ~ total_bill', data=data).fit()

# Obtendo os resultados da regressão
coeficiente_angular = regression_model.params['total_bill']
intercepto = regression_model.params['Intercept']

# Imprimindo os resultados da regressão
print("Coeficiente Angular:", coeficiente_angular)
print("Intercepto:", intercepto)

# Reproduzindo o gráfico de dispersão
plt.scatter(data['total_bill'], data['tip'])
plt.plot(data['total_bill'], regression_model.fittedvalues, color='red', linewidth=2)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Gráfico de Dispersão de Tip vs. Total Bill')
plt.show()

````

### 2. Mudança de perspectiva

Agora ajuste um modelo de regressão de ```tip_pct``` por ```net_bill```. Construa o gráfico de dispersão correspondente.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Carregando os dados usando o seaborn
data = sns.load_dataset('tips')

# Criando a coluna 'tip_pct', que representa a porcentagem da gorjeta em relação à conta total
data['tip_pct'] = data['tip'] / data['total_bill'] * 100

# Ajustando o modelo de regressão linear usando statsmodels
regression_model = smf.ols('tip_pct ~ total_bill', data=data).fit()

# Obtendo os resultados da regressão
coeficiente_angular = regression_model.params['total_bill']
intercepto = regression_model.params['Intercept']

# Imprimindo os resultados da regressão
print("Coeficiente Angular:", coeficiente_angular)
print("Intercepto:", intercepto)

# Reproduzindo o gráfico de dispersão
plt.scatter(data['total_bill'], data['tip_pct'])
plt.plot(data['total_bill'], regression_model.fittedvalues, color='red', linewidth=2)
plt.xlabel('Total Bill')
plt.ylabel('Tip Percentage')
plt.title('Gráfico de Dispersão de Tip Percentage vs. Total Bill')
plt.show()
````

Compare os dois gráficos e comente (de forma objetiva e suscinta) a relação entre eles e a diferença do ajuste. Podemos comparar diretamente os $R^2$s?

O gráfico de dispersão entre tip_pct e net_bill mostra uma relação linear positiva entre essas duas variáveis. À medida que o valor de net_bill aumenta, a porcentagem da gorjeta (tip_pct) também aumenta, indicando que a gorjeta tende a aumentar em proporção à conta total.

O gráfico de dispersão entre tip e net_bill mostra uma relação linear positiva mais fraca em comparação com o gráfico anterior. A diferença no ajuste entre os dois modelos é refletida pelos valores de $R^2$. O $R^2$ é uma medida que indica a proporção da variabilidade da variável

### 4. Ajuste do modelo em outra escala

Com o modelo no item 3, obtenha a preditiva do valor da gorjeta:
